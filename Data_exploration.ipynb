### Associations between features

```python
corrmat = df.corr()
f, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(corrmat, cmap='RdYlGn');
```

```python
sns.set()
features = df.copy()
features = features.drop(['Outcome'],axis=1)
xvars = features.columns
sns.pairplot(df, y_vars=['Outcome'],x_vars=xvars[:8]) #Verander de :8 naarmate je meer of minder variabelen wilt bekijken
plt.show();
```
